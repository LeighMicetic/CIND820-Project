```python
#removes any response that contains refused

#combines variables into one file

combine_file = X_newnames.copy()
combine_file["Number_of_Doctors_Visited"] = y_new

#sets the value we want removed

value_to_remove = 'Refused'

#creates a mask of the rows to keep after taking out rows with 'Refused'

mask = ~combine_file.apply(lambda row: (row == value_to_remove).any(), axis=1)

#creates the new file

filtered_file = combine_file[mask]

#separates the new file back into a file for independent variables and the dependent variable

X_filtered = filtered_file.drop(columns = 'Number_of_Doctors_Visited')
y_filtered = filtered_file['Number_of_Doctors_Visited']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#creates a function to plot a relative frequency plot for each independent variable

def relative_frequency_plot(df):

  for column in df.columns:
    if df[column].dtype == 'category':
      relative_frequencies = df[column].value_counts(normalize=True)
      plt.figure(figsize=(3, 2))
      sns.barplot(x=relative_frequencies.index, y=relative_frequencies.values)
      plt.title(f'Relative Frequency Plot for {column}')
      plt.xlabel(column)
      plt.ylabel('Relative Frequency')
      plt.xticks(rotation=45)
      plt.show()

relative_frequency_plot(X_filtered)

#creates a frequency plot for the dependent variable

frequency = y_filtered.value_counts(normalize=True)
plt.figure(figsize=(3, 2))
frequency.plot(kind = 'bar')
plt.title('Relative Frequency Plot for Number of Doctors Visited')
plt.xlabel('Number of Doctors Visited')
plt.ylabel('Realtive Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
#split into train-test sets

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

features = pd.DataFrame(X_filtered)
target = pd.DataFrame(y_filtered)

#independent variables and dependent variables

X = features
y = target

#encodes the categorical independent variables using OneHotEncoder

encoder = OneHotEncoder(sparse=False, drop='first')
X_encoded = encoder.fit_transform(X)

#encodes the categorical dependent variable using LabelEncoder

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

#ensures y is a 1D array

y_encoded = y_encoded.ravel()

#splits the data into training and test sets with a test size of 20%

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size = 0.2, random_state = 42)

#confirm test size

print("There are", len(X_train), "observations in the training set")
print("There are", len(X_test), "observations in the test set")

```

```python
#Logistic Regression

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import cross_val_score
import time

#initialize the logistic regression model

logreg = LogisticRegression(max_iter=1000)

#records the start time of the algorithm

logreg_time = time.time()

#fits the model to the training data

logreg.fit(X_train, y_train)

#Make prediction using the test set

lr_pred = logreg.predict(X_test)

#records the end time of the logistic regression prediction model

logreg_end = time.time()

#Calculates the difference between the start time and end time of the algorithm

logreg_elapsed = logreg_end - logreg_time

#prints out the time it took to run the algorithm

print(f"Time taken to perfom logistic regression: {logreg_elapsed:.4f} seconds")
print("\n")
#creates the label names for the Classification Report

target_names = ['0-1 doctors', '2-3 doctors', '4 or more doctors']

print("Logistic Regression Classification Report")
print(classification_report(y_test, lr_pred, target_names=target_names))

print("Logistic Regression Confusion Matrix")
cm = confusion_matrix(y_test, lr_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels = label_encoder.classes_)

plt.figure(figsize=(8,6))
disp.plot(cmap=plt.cm.Blues)

```

The Logistic Regression model took 0.1192 seconds to be run and was able to correctly predict the number of doctors that an individual visits 33% of the time for those seeing 0-1 doctors, 56% of the time for those seeing 2-3 doctors and 36% of the time for those seeing 4 or more doctors. Of the predictions made, only 5% were correct for those individuals visting 0-1 doctors, 82% correct for those visitng 2-3 doctors and 21% correct for those visiting 4 or more doctors.  The overall accuracy of the Logistic Regression model was 52%.  

```python
#Decision Tree

from sklearn.tree import DecisionTreeClassifier, plot_tree

#Trains the model using the training sets

decision = DecisionTreeClassifier(random_state=42)

#records the start time of the algorithm

decision_start = time.time()

#fits the model to the training data

decision.fit(X_train, y_train)

#make predictions using the test set

dt_pred = decision.predict(X_test)

#records the end time of the prediction model

decision_end = time.time()

#Calculates the difference between the start time and end time of the algorithm

decision_elapsed = decision_end - decision_start

#prints out the time it took to run the algorithm

print(f"Time taken to perform decision tree model: {decision_elapsed:.4f} seconds")
print("\n")
print("Decision Tree Classification Report")
print(classification_report(y_test, dt_pred, target_names=target_names))

print("Decision Tree Confusion Matrix")
cm = confusion_matrix(y_test, dt_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels = label_encoder.classes_)

plt.figure(figsize=(8,6))
disp.plot(cmap=plt.cm.Blues)
```

The Decision Tree model took 0.0071 seconds to be run and was able to correctly predict the number of doctors that an individual visits 10% of the time for those seeing 0-1 doctors, 57% of the time for those seeing 2-3 doctors and 28% of the time for those seeing 4 or more doctors.  Of the predictions made, 14% were correct for those individuals visiting 0-1 doctors, 47% correct for those visiting 2-3 doctors and 31% correct for those visiting 4 or more doctors.  The overall accuracy of the Decision Tree model was 37%.

```python
#Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier

#Train the random forest model

random = RandomForestClassifier(random_state=42)

#records the start time of the algorithm

random_start = time.time()

#fits the model to the training data
random.fit(X_train, y_train)

#make predictions using the test set

rf_pred = random.predict(X_test)

#records the end time of the prediction model

random_end = time.time()

#Calculates the difference between the start time and end time of the algorithm

random_elapsed = random_end - random_start

#prints out the time it took to run the algorithm

print(f"Time taken to perform random forest prediction model: {random_elapsed:.4f} seconds")
print("\n")
print("Random Forest Classification Report")
print(classification_report(y_test, rf_pred, target_names=target_names))

print("Random Forest Confusion Matrix")
cm = confusion_matrix(y_test, rf_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels = label_encoder.classes_)

plt.figure(figsize=(8,6))
disp.plot(cmap=plt.cm.Blues)
```

The Random Forest model tool 0.3782 seconds to be run and was able to correctly predict the number of doctors that an individual visits 14% of the time for those seeing 0-1 doctors, 59% of the time for those seeing 2-3 doctors and 30% of the time for those seeing 4 or more doctors.  Of the predictions made, 14% were correct for those individuals visiting 0-1 doctors, 65% correct for those visiting 2-3 doctors and 24% correct for those visiting 4 or more doctors.  The overall accuracy of the Random Forest model was 45%.

Out of the three models, the Logistic Regression model is considered to be the better model. It had the best accuracy score (52%), was more evenly precise (33%, 56% and 36%) and had the highest recall for one of the categories (2-3 doctors at 82%) than the other models although it was not the fastest model to be run.   

```python
#Cross-Validation of Training Set, Test set

#creates a function to produce the cross validation scores for each of the models on the training and test sets separately

models = {"Logistic Regression": logreg, "Decision Tree": decision, "Random Forest": random}

for name, model in models.items():
  cv_scores = cross_val_score(model, X_train, y_train, cv=10)

  print(f"\n{name} Cross-Validation Scores (Training Set):", cv_scores)
  print(f"\n{name} Mean Cross_Validation Scores (Training Set):", np.mean(cv_scores))

for name, model in models.items():
  cv_scores = cross_val_score(model, X_test, y_test, cv=10)

  print(f"\n{name} Cross-Validation Scores (Test Set):", cv_scores)
  print(f"\n{name} Mean Cross_Validation Scores (Test Set):", np.mean(cv_scores))
```

Cross-Validation with 10 folds of the training and test sets was conducted to determine the stability of the prediciton models.  The mean score of the cross-validation of Logistic Regression model for the training set was 48% and for the test set 46%.  The mean score of the cross_validation for the Decision Tree model for the training set was 38% and for the test set 32%.
