```python
#removes any response that contains refused

#combines variables into one file

combine_file = X_newnames.copy()
combine_file["Number_of_Doctors_Visited"] = y_new

#sets the value we want removed

value_to_remove = 'Refused'

#creates a mask of the rows to keep after taking out rows with 'Refused'

mask = ~combine_file.apply(lambda row: (row == value_to_remove).any(), axis=1)

#creates the new file

filtered_file = combine_file[mask]

#separates the new file back into a file for independent variables and the dependent variable

X_filtered = filtered_file.drop(columns = 'Number_of_Doctors_Visited')
y_filtered = filtered_file['Number_of_Doctors_Visited']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#creates a function to plot a relative frequency plot for each independent variable

def relative_frequency_plot(df):

  for column in df.columns:
    if df[column].dtype == 'category':
      relative_frequencies = df[column].value_counts(normalize=True)
      plt.figure(figsize=(3, 2))
      sns.barplot(x=relative_frequencies.index, y=relative_frequencies.values)
      plt.title(f'Relative Frequency Plot for {column}')
      plt.xlabel(column)
      plt.ylabel('Relative Frequency')
      plt.xticks(rotation=45)
      plt.show()

relative_frequency_plot(X_filtered)

#creates a frequency plot for the dependent variable

frequency = y_filtered.value_counts(normalize=True)
plt.figure(figsize=(3, 2))
frequency.plot(kind = 'bar')
plt.title('Relative Frequency Plot for Number of Doctors Visited')
plt.xlabel('Number of Doctors Visited')
plt.ylabel('Realtive Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
#split into train-test sets

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

features = pd.DataFrame(X_filtered)
target = pd.DataFrame(y_filtered)

#independent variables and dependent variables

X = features
y = target

#encodes the categorical independent variables using OneHotEncoder

encoder = OneHotEncoder(sparse=False, drop='first')
X_encoded = encoder.fit_transform(X)

#encodes the categorical dependent variable using LabelEncoder

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

#ensures y is a 1D array

y_encoded = y_encoded.ravel()

#splits the data into training and test sets with a test size of 20%

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size = 0.2, random_state = 42)

#confirm test size

print("There are", len(X_train), "observations in the training set")
print("There are", len(X_test), "observations in the test set")

```

```python
#Logistic Regression

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import cross_val_score

#Train the model with the training set

logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)

#Make prediction using the test set

lr_pred = logreg.predict(X_test)

#creates the label names for the Classification Report

target_names = ['0-1 doctors', '2-3 doctors', '4 or more doctors']

```

```python
#Decision Tree

from sklearn.tree import DecisionTreeClassifier, plot_tree

#Trains the model using the training sets

decision = DecisionTreeClassifier(random_state=42)
decision.fit(X_train, y_train)

#make predictions using the test set

dt_pred = decision.predict(X_test)

```

```python
#Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier

#Train the random forest model

random = RandomForestClassifier(random_state=42)
random.fit(X_train, y_train)

#make predictions using the test set

rf_pred = random.predict(X_test)

```

```python
#Comparison of Logistic Regression, Decision Tree and Random Forest

#Initialize Models

models ={
    "Logistic Regression": logreg,
    "Decision Tree": decision,
    "Random Forest": random,
}

#train and evaluate models

for name, model in models.items():
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)

  print(f"\n{name} Classification Report")
  print(classification_report(y_test, y_pred, target_names=target_names))

  #Confusion Matrix

  cm = confusion_matrix(y_test, y_pred)
  disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels = label_encoder.classes_)

  plt.figure(figsize=(8,6))
  disp.plot(cmap=plt.cm.Blues)
  plt.title(f"{name} Confusion Matrix")
  plt.show()

  #Cross-Validation Score
  cv_scores = cross_val_score(model, X_encoded, y_encoded, cv=10)
  print(f"{name} Cross_Validation Scores:", cv_scores)
  print(f"{name} Mean Cross_Validation Scores", np.mean(cv_scores))
```

```python
#Cross-Validation of Training Set, Test set

for name, model in models.items():
  cv_scores = cross_val_score(model, X_train, y_train, cv=5)

  print(f"\n{name} Cross-Validation Scores (Training Set):", cv_scores)
  print(f"\n{name} Mean Cross_Validation Scores (Training Set):", np.mean(cv_scores))

for name, model in models.items():
  cv_scores = cross_val_score(model, X_test, y_test, cv=10)

  print(f"\n{name} Cross-Validation Scores (Test Set):", cv_scores)
  print(f"\n{name} Mean Cross_Validation Scores (Test Set):", np.mean(cv_scores))
```

```python
#The accuracy scores of the prediction models using Logistic Regression, Decision Tree and Randforest Classifier were 52%, 37% and 45%.
#As none of the models performed well using all the features available, a chi square analysis will be done to select the best features and the models will be run again

#chi square to determine best features

from sklearn.feature_selection import chi2

#compute Chi-Square scores

chi2_scores, p_values = chi2(X_encoded, y_encoded)
feature_names = encoder.get_feature_names_out(features.columns)
chi2_results = pd.DataFrame({'Feature': feature_names, 'Chi2 Score': chi2_scores, 'p-value': p_values})

print(chi2_results.sort_values(by='Chi2 Score', ascending = False))
```

```python
#From the Chi-Square anaylsis we will filter the features being used to those with a p-value less 0.05.
#Features will include:  Sleep_Meds, Medication_Sleeping, Physical_Health, Race and Mental Health
#Analysis will be run again but only using these features.

#independent variables and dependent variables

filtered_features = features[['Physical_Health', 'Employment', 'Medication_Sleeping', 'Race', 'Sleep_Meds']]

#encodes the categorical independent variables using OneHotEncoder

encoder = OneHotEncoder(sparse=False, drop='first')
X_encoded = encoder.fit_transform(filtered_features)

#encodes the categorical dependent variable using LabelEncoder

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

#ensures y is a 1D array

y_encoded = y_encoded.ravel()

#splits the data into training and test sets with a test size of 20%

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size = 0.2, random_state = 42)

#confirm test size

print("There are", len(X_train), "observations in the training set")
print("There are", len(X_test), "observations in the test set")
```

```python
#Comparison of Logistic Regression, Decision Tree and Random Forest with filtered features

#Initialize Models

models ={
    "Logistic Regression": logreg,
    "Decision Tree": decision,
    "Random Forest": random,
}

#train and evaluate models

for name, model in models.items():
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)

  print(f"\n{name} Classification Report")
  print(classification_report(y_test, y_pred, target_names=target_names))

  #Confusion Matrix

  cm = confusion_matrix(y_test, y_pred)
  disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels = label_encoder.classes_)

  plt.figure(figsize=(8,6))
  disp.plot(cmap=plt.cm.Blues)
  plt.title(f"{name} Confusion Matrix")
  plt.show()

  #Cross-Validation Score
  cv_scores = cross_val_score(model, X_encoded, y_encoded, cv=5)
  print(f"{name} Cross_Validation Scores:", cv_scores)
  print(f"{name} Mean Cross_Validtion Scores", np.mean(cv_scores))
```
